```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.colors as mcolors
import pickle

import glob
from PIL import Image
from scipy import interpolate
import warnings

warnings.filterwarnings('ignore')
```

```python
# Get all the csv files in the current directory
csv_files = glob.glob("../../experiments/logs/simdata/fig2/volterra/exp_*.csv")

# Create a list of dataframes
dfs = []

# Iterate over the csv files
for csv_file in csv_files:

    # Read the csv file into a dataframe
    df = pd.read_csv(csv_file)

    # Append the dataframe to the list of dataframes
    dfs.append(df)

# Concatenate the dataframes into a single dataframe
df = pd.concat(dfs)
```

```python
num_epochs = 2000
coeffs, other_columns = [], []

for i in range(3):
        for j in range(3):
            for k in range(3):
                coeffs.append(f"A_{i}{j}{k}")

for column in df.columns:
    if column not in set(coeffs):
        other_columns.append(column)
print(other_columns)

for coeff in coeffs:
    df[coeff] = pd.to_numeric(df[coeff])
df["moving_avg_window"] = df["moving_avg_window"].astype(int)
df["trials_per_block"] = df["trials_per_block"].astype(int)
df["flyid"].unique()
```

```python
l1_reg = 1e-2
df_subset = df.loc[(df['epoch'] == num_epochs) & (df['l1_regularization'] == l1_reg)]

for coeff in coeffs:
    if df_subset[coeff].sum() == 0:
        df_subset = df_subset.drop(coeff, axis=1)

df_plot = df.loc[(df['l1_regularization'] == l1_reg) & (df['epoch'] <= num_epochs) & (df['flyid'] == 3)]
df_plot2 = df.loc[(df['l1_regularization'] == 1e-2) & (df['epoch'] == num_epochs)]

df_subset.head()
```

```python
# Get all the csv files in the current directory
csv_files = glob.glob("../../experiments/logs/simdata/fig2/mlp/exp_*.csv")

# Create a list of dataframes
dfs = []

# Iterate over the csv files
for csv_file in csv_files:

    # Read the csv file into a dataframe
    temp_df = pd.read_csv(csv_file)

    # Append the dataframe to the list of dataframes
    dfs.append(temp_df)

# Concatenate the dataframes into a single dataframe
mlp_df = pd.concat(dfs)


mlp_df_plot2 = mlp_df.loc[(mlp_df['epoch'] == num_epochs)]

common_columns = ["percent_deviance", "plasticity_model", "r2_weights"]
df1_common = df_plot2[common_columns]
df2_common = mlp_df_plot2[common_columns]

# Append the dataframes
joint_df = pd.concat([df1_common, df2_common], ignore_index=True)
```

```python
melted_df = pd.melt(df_subset, id_vars=other_columns, var_name="coeff_name", value_name="coeff_val").astype({'coeff_val': 'float'})
```

```python
sns.set_theme(style="whitegrid")
sns.set(font_scale = 1.2)
sns.set_style("white")

fig, axs = plt.subplots(nrows=1, ncols=4,  figsize=(15, 5), width_ratios=[3,3,1.5,1.5])

# fig.suptitle('Fit Volterra Coefficients', fontsize="medium")
            
axs[1].axhline(y=0, color='r', linestyle='--')
sns.boxplot(data=melted_df, x="coeff_name", y="coeff_val", ax=axs[1], width=1)
axs[1].set_xlabel("Plasticity coefficients", fontsize=13)
axs[1].set_ylabel(None)
axs[1].tick_params(labelrotation=90, which="major", labelsize=14, color="black", direction='out')

axs[2].set_title('% Dev Explained')
sns.boxplot(data=joint_df, x="plasticity_model", y="percent_deviance", hue="plasticity_model", width=0.3, ax=axs[2])
axs[2].set_ylabel(None)
axs[2].set_xlabel(None)

axs[3].set_title('R2 score')
sns.boxplot(data=joint_df, x="plasticity_model", y="r2_weights", hue="plasticity_model", width=0.3, ax=axs[3])
axs[3].legend().remove()
axs[3].set_ylabel(None)
axs[3].set_xlabel(None)

for i in range(3):
    for j in range(3):
        for k in range(3):    
            col = '#625D5D'
            if i == 1 and j == 1 and k == 0:
                # col = '#4863A0'
                col = 'coral'
            
            sns.lineplot(x=df_plot["epoch"], y=df_plot['A_'+str(i)+str(j)+str(k)], color = col, ax=axs[0])

axs[0].set_xlabel('Epochs', fontsize=13)
axs[0].set_ylabel('$A_{ijk}$', fontsize=13)

plt.tight_layout()
plt.savefig("fig2-plasticity-fit.svg", dpi=500)
```
